```python
import matplotlib.pyplot as plt
```

```python
with open("../checkpoints/exp2/train.log", "r") as f:
    data = f.readlines()
```

```python
data = [line.strip() for line in data if "Step" in line]
data[:4]
```

```python
# steps
steps = [int(line.split("|")[1].strip().replace("Step ", "")) for line in data]

# mel loss
mel_loss = [float(line.split("|")[6].strip().replace("Mel ", "")) for line in data]
mel_loss[:4]
```

```python
plt.title("Mel loss")
plt.plot(steps, mel_loss)
plt.show()
```
